```python
#The code below is directly from https://github.com/DemisEom/SpecAugment.
#just for computational cost's optimization, i deconstructed each of them.
#No copyright infrienge intended
'''
/usr/local/lib/python3.7/dist-packages/specAugment/spec_augment_tensorflow.py
overwrite this with shared spec_augment_tensorflow.py for deconstruction of augmentation policies.
'''
%tensorflow_version 1.x
import argparse
import librosa
from specAugment import spec_augment_tensorflow
import os, sys
import numpy as np
from pedalboard import Pedalboard, Reverb
from pedalboard.io import AudioFile
import soundfile as sf
import pyroomacoustics as pra
def roomacaudio(audio, sampling_rate):
  # The desired reverberation time and dimensions of the room
  rt60 = 0.5  # seconds
  room_dim = [9, 7.5, 3.5]  # meters
  # We invert Sabine's formula to obtain the parameters for the ISM simulator
  e_absorption, max_order = pra.inverse_sabine(rt60, room_dim)
  # Create the room
  room = pra.ShoeBox(
      room_dim, fs=sampling_rate, materials=pra.Material(e_absorption), max_order=max_order
  )
  # place the source in the room
  room.add_source([2.5, 3.73, 1.76], signal=audio, delay=1.3)
  # define the locations of the microphones
  mic_locs = np.c_[
      [6.3, 4.87, 1.2],  # mic 1
  ]

  # finally place the array in the room
  room.add_microphone_array(mic_locs)

  room.simulate()
  cutoff = np.where(room.mic_array.signals[0,:]>room.mic_array.signals[0,:][0])[0][0]
  y = room.mic_array.signals[0,:][cutoff:cutoff+np.shape(audio)[0]]
  return y
```

```python
audio_path = "/content/drive/MyDrive/LibriSpeech/LibriSpeech/dev-clean/1272/128104/1272-128104-0000.flac"
# If you are Pytorch, then import spec_augment_pytorch instead of spec_augment_tensorflow
audio, sampling_rate = librosa.load(audio_path)
#each of these
mel_spectrogram = librosa.feature.melspectrogram(y=audio,
                                                    sr=sampling_rate,
                                                    n_mels=256,
                                                    hop_length=128,
                                                    fmax=8000)
warped_masked_spectrogram = spec_augment_tensorflow.freqmask(mel_spectrogram=mel_spectrogram)
spec_augment_tensorflow.visualization_spectrogram(warped_masked_spectrogram, title="freqmask")

warped_masked_spectrogram = spec_augment_tensorflow.timewarp(mel_spectrogram=mel_spectrogram)
spec_augment_tensorflow.visualization_spectrogram(warped_masked_spectrogram, title="timewarp")

warped_masked_spectrogram = spec_augment_tensorflow.timemask(mel_spectrogram=mel_spectrogram)
spec_augment_tensorflow.visualization_spectrogram(warped_masked_spectrogram, title="timemask")


# Make a Pedalboard object, containing multiple plugins:
# board = Pedalboard([Reverb(room_size=0.25)])
board = Pedalboard([Reverb(room_size=0.8)])

# Run the audio through this pedalboard!
effected = board(audio, sampling_rate)
reverbed = librosa.feature.melspectrogram(y=effected,
                                          sr=sampling_rate,
                                          n_mels=256,
                                          hop_length=128,
                                          fmax=8000)

spec_augment_tensorflow.visualization_spectrogram(reverbed, title="reverb")

roomac = librosa.feature.melspectrogram(y=roomacaudio(audio, sampling_rate),
                                          sr=sampling_rate,
                                          n_mels=256,
                                          hop_length=128,
                                          fmax=8000)
spec_augment_tensorflow.visualization_spectrogram(roomac, title="roomac")
```
